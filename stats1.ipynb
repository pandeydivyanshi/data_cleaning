statistical analysis of clean data set

```python
import pandas as pd

# Load cleaned data
df = pd.read_csv(r"C:\Users\divya\Desktop\data_cleaning\data_cleaning\cleaned_file.csv")
```

```python
DESCRIPTIVE_STATS = df.describe().round(3)  # Descriptive statistics
```

```python
print("ðŸ”¹ Descriptive Statistics:")
print(df.describe().round(3))                        ##it tells about the data like mean, std, min, max, 25%, 50%, 75% etc.
```

checking missing value for safty perpose

```python
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())
```

SKEWNESS AND KURTOSIS

```python
print("\nðŸ”¹ Skewness:")
print(df.skew(numeric_only=True).round(3))

print("\nðŸ”¹ Kurtosis:")
print(df.kurt(numeric_only=True).round(3))
```

CORRELATION

```python
print("\nðŸ”¹ Correlation Matrix:")
print(df.corr(numeric_only=True).round(3))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")              #WHITE GRID STYLE
for col in df.select_dtypes(include='number').columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue', bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
selected = df.select_dtypes(include='number').columns[:4]  # First 4 numeric columns
sns.pairplot(df[selected], diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
