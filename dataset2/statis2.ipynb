```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x=pd.read_csv(r'C:\Users\divya\OneDrive\Desktop\data_cleaning\data_cleaning\dataset2\cleaned_file2.csv')
```

#descriptive statisctics

```python
print("ðŸ”¹ Descriptive Statistics:")
print(x.describe().round(3))
```

#for saftey check empty cells

```python
print("\nðŸ”¹ Missing Values:")
print(x.isnull().sum())
```

#SKEWNESS AND KURTOSIS

```python
print("\nðŸ”¹ Skewness:")
print(x.skew(numeric_only=True).round(3))

print("\nðŸ”¹ Kurtosis:")
print(x.kurt(numeric_only=True).round(3))
```

#correlation

```python
print("\nðŸ”¹ Correlation Matrix:")
print(x.corr(numeric_only=True).round(3))
```

#visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")              #WHITE GRID STYLE
for col in x.select_dtypes(include='number').columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(x[col], kde=True, color='skyblue', bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

#pair plot

```python
selected = x.select_dtypes(include='number').columns[:4]  # First 4 numeric columns
sns.pairplot(x[selected], diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
