#loading original data

```python

import pandas as pd
import numpy as np
df=pd.read_csv(r'c:\Users\divya\Downloads\numeric_data_cleaning.csv')
x=df.copy()                                    ##copying the variable
print(x.to_string())
```

#basic info about data

```python
print("Dataset Shape:", x.shape)
print(f"number of rows: {len(x)}")
print("Column Names:\n", x.columns.tolist())
print(x.index)
print(x.info())
print(x.head())
```

#working on empty cells


```python
print(x.isnull().sum())  # Count of null values in each column
print(x.isna().sum())   # Count of null values in each column (alternative method)
print("Filling null values with mean...")
x['Income'].fillna(round(x['Income'].mean(), 2), inplace=True)
x['Score'].fillna(round(x['Score'].mean(), 2), inplace=True)
print("Null values after filling with mean:")
print(x.isnull().sum())  # Count of null values after filling
```

#working on wrong formate


```python
print(x.dtypes)
print(x.info())             
##no Object = no wrong formate
##no wrong formate in this dataset
```

#checking for wrong data

```python
# Show first few rows
print(x.head())

# ------- 1. Check for negative values -------
print("\nüîç Checking for negative values:")
neg_indices = np.where(x < 0)
print(neg_indices)

# Example: Replace negative Age at row 100 with average Age
if 100 in neg_indices[0] and 'Age' in x.columns:
    mean_age = int(df['Age'].mean())
    x.loc[100, 'Age'] = mean_age
    print(f"\n‚úÖ Replaced negative Age at row 100 with mean Age: {mean_age}")

# ------- 2. Check for extreme outliers -------
print("\nüîç Checking extreme Purchases (should not be > 20):")
outlier_indices = np.where(x['Purchases'] > 20)
print(outlier_indices)

# Example: Replace outlier in Purchases at row 150 with average purchase value (say 10)
if 150 in outlier_indices[0]:
    x.loc[150, 'Purchases'] = 10
    print(f"\n‚úÖ Replaced Purchases at row 150 with 10")

# ------- 3. Final check -------
print("\nüßº Cleaned Data Preview:\n", x.head())
```

#column wise data type
. changing score in int dtype
. income in int dtype with rupee

```python

#1. Convert 'Score' to integer type
x['Score'] = x['Score'].astype(int)

# 2. Rename column 'Income' to 'Income in Rs.'
x= x.rename(columns={'Income': 'Income in Rs.'})

# 3. Convert 'Income in Rs.' to integer type
x['Income in Rs.'] = x['Income in Rs.'].astype(int)
print(x.to_string())
```

# dealing with duplicate rows

```python


# -------- 1. Check for duplicate rows --------
print("\nüîç Number of duplicate rows:", x.duplicated().sum())

# -------- 2. View duplicate rows (optional) --------
print("\nüìã Duplicate Rows:")
print(x[x.duplicated()])

# -------- 3. Remove duplicate rows --------
x.drop_duplicates(inplace=True)

# -------- 4. Confirm removal --------
print("\n‚úÖ After removing duplicates, new shape:", x.shape)
```
