The aim of this project is to clean, process, analyze, and visualize a numeric dataset containing columns such as Age, Income, Score, and Purchases. The final goal is to prepare this dataset for further statistical modeling and decision-making.
To perform data cleaning on a numerical dataset by:

Treating outliers

Correcting data types and labels

```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r'c:\Users\divya\Downloads\numeric_data_cleaning.csv')
x=df.copy()                                    ##copying the variable
x.head()
```

üß≠ Data Overview

```python
print("Shape:", x.shape)
print("Columns:", x.columns.tolist())
print(x.dtypes)
print(x.isnull().sum())
```

üßπ Data Cleaning Summary
‚úÖ Steps Taken:
Filled missing values in 'Income' and 'Score' with their mean.

Converted all values to positive using .abs() where logical.

Replaced extreme outliers in 'Purchases' (values > 20).

Renamed 'Income' to 'Income in Rs.'.

Converted 'Score' and 'Income in Rs.' to int.

Removed duplicate rows using drop_duplicates().

```python
x.isnull().sum()
x['Income'].fillna(round(x['Income'].mean(), 2), inplace=True)
x['Score'].fillna(round(x['Score'].mean(), 2), inplace=True)
neg_indices = np.where(x < 0)
x.loc[100, 'Age'] = int(df['Age'].mean())
outlier_indices = np.where(x['Purchases'] > 20)
x['Score'] = x['Score'].astype(int)
x.rename(columns={'Income': 'Income in Rs.'})
x.duplicated().sum()
```

üìä  Statistical Summary
Summary statistics provided:

Mean, Median, Std Dev, Min, Max, IQR for all numeric fields.

Useful for identifying skewed distributions and validating cleaning steps.

skeweness,kurtosis,correlation

```python
x.describe().round(3)
print("Skewness:\n", x.skew(numeric_only=True).round(3))
print("\nKurtosis:\n", x.kurt(numeric_only=True).round(3))
plt.figure(figsize=(10, 8))
sns.heatmap(x.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

üìà  Visualizations
üß™ Histograms
Distribution of numeric variables to detect skewness or gaps in data

```python
for col in x.select_dtypes(include='number').columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(x[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

#pair plot

```python
sns.pairplot(x[x.columns[:4]], diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
